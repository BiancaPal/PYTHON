# Scatter Plots

Bokeh can draw many types of visual shapes (Called glyphs), including lines, bars, patches, hex lines and more. One of the most common visualization tasks is to draw a scatter plot of data using small marker glyphs to represent each point.

In this section you will see how to use Bokeh's various marker glyphs to create simple scatter plots.

with p= figure(...) => we create a blank figure
to call a glyph method on the figure => p.circle
show(p) => to show the figure




```python
# create a new plot with default tools, using figure
p = figure(plot_width=400, plot_height=400)

# add a circle renderer with x and y coordinates, size, color, and alpha (intensity of the color)
p.circle([1, 2, 3, 4, 5],[3, 7, 4, 4, 5], radius=0.10, line_color="navy", fill_color="blue", fill_alpha=1)

show(p) #show the results
```

```python
p = figure(plot_width=400,plot_height=400)

# in here we use the squared method, and instead of giving a general size, we give each marker a different one, and also color as a shorthand to set both the line and fill colors easily at the same time
p.square([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=[10, 15, 20, 25, 30], color ="firebrick", alpha=0.9)

show(p)
```

# Line Plots

```python
p = figure(plot_width = 400, plot_height=400, title="My line Plot")

p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2 )

show(p)
```

```python
from bokeh.sampledata.glucose import data
data.head()
```

```python
# reduce data size to week
week = data.loc['2010-10-01':'2010-10-08']

p = figure(x_axis_type='datetime',title="Glucose Range", plot_height=350, plot_width=800)
p.xgrid.grid_line_color=None
p.ygrid.grid_line_alpha=0.5
p.xaxis.axis_label= 'Time'
p.yaxis.axis_label= 'Value'

p.line(week.index, week.glucose)

show(p)
```

```python
# EXERCISE: Look at the AAPL, FB, GOOGL, IBM, MSFT data from bokeh.sampledata.stocks and create a line plot using it

from bokeh.sampledata.stocks import AAPL, FB, GOOG, IBM, MSFT 
      
# instantiating the figure object 
graph = figure(x_axis_type = "datetime", title = "Stock Closing Prices") 
   
# name of the x-axis 
graph.xaxis.axis_label = 'Date'
   
# name of the y-axis 
graph.yaxis.axis_label = 'Price (in USD)'
   
# plotting the line graph for AAPL 
x_axis_coordinates = np.array(AAPL['date'], dtype = np.datetime64) 
y_axis_coordinates = AAPL['adj_close'] 
color = "lightblue"
legend_label = 'AAPL'
graph.line(x_axis_coordinates, 
        y_axis_coordinates, 
        color = color, 
        legend_label = legend_label) 
   
# plotting the line graph for FB 
x_axis_coordinates = np.array(FB['date'], dtype = np.datetime64) 
y_axis_coordinates = FB['adj_close'] 
color = "black"
legend_label = 'FB'
graph.line(x_axis_coordinates, 
        y_axis_coordinates, 
        color = color, 
        legend_label = legend_label) 
   
# plotting the line graph for GOOG 
x_axis_coordinates = np.array(GOOG['date'], dtype = np.datetime64) 
y_axis_coordinates = GOOG['adj_close'] 
color = "orange"
legend_label = 'GOOG'
graph.line(x_axis_coordinates, 
        y_axis_coordinates, 
        color = color, 
        legend_label = legend_label) 
   
# plotting the line graph for IBM 
x_axis_coordinates = np.array(IBM['date'], dtype = np.datetime64) 
y_axis_coordinates = IBM['adj_close'] 
color = "darkblue"
legend_label = 'IBM'
graph.line(x_axis_coordinates, 
        y_axis_coordinates, 
        color = color, 
        legend_label = legend_label) 
   
# plotting the line graph for MSFT 
x_axis_coordinates = np.array(MSFT['date'], dtype = np.datetime64) 
y_axis_coordinates = MSFT['adj_close'] 
color = "yellow"
legend_label = 'MSFT'
graph.line(x_axis_coordinates, 
        y_axis_coordinates, 
        color = color, 
        legend_label = legend_label) 
   
# relocating the legend table to  
# avoid abstruction of the graph 
graph.legend.location = "top_left"
   
# displaying the model 
show(graph) 
```
