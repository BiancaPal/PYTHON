# Cleaning Data

Problem 1: The length should be a number, not a string
Problem 2: The title of every movie has the year in parenthesis ( it's redundant)
Problem 3: The genre is a string, it could be better if it was a list, if we could split this we could to a better data analysis


```python
data['formatted title'] = data['title'].str[:-7]
```

```python
data.head()
```

```python
data['formatted title'] = data['title'].str.split(' \(').str[0]
```

```python
data.head()
```

```python
data['length'].str[:-6]
```

```python
data['formatted length'] = data['length'].str.replace(' mins.','').astype('int')
```

```python
sns.pairplot(data)
```

```python
data[data['formatted length'] == 0]
```

```python
data['formatted length'] = data['length'].str.split().str.get(0).astype('int')
```

```python
sns.pairplot(data)
```

```python
data[data['formatted length'] == 0]
```

```python
data['genre type 1'] = data['genres'].str.split('|').str[0]
```
