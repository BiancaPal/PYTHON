# Making a vertical bar chart

```python
df=pd.read_csv('gdp_per_capita.csv')
```

```python
plt.bar(df.YEAR, df['PIB_per_capita'])
plt.ylabel("GDP per capita")
plt.xlabel('YEAR')
plt.show()
```

# Making a horizontal bar chart

```python
plt.barh(df.YEAR, df['PIB_per_capita'])
plt.ylabel("GDP per capita")
plt.xlabel('YEAR')
plt.show()
```

# Adding error bars


```python
df=pd.read_csv('gdp_per_field.csv')
plt.bar(df.name, df.average, yerr=df.deviation)
plt.ylabel('GDP per fiels')
plt.show()
```

# Stacked bar charts

```python
df=pd.read_csv('occupation_broken_into_sex.csv')
```

```python
print(df.TRIMESTER, df['FEMALE(in thousands)'],df['MALE(in thousands)'])
```

```python
plt.bar(df.TRIMESTER, df['MALE(in thousands)'],label='Male')
plt.bar(df.TRIMESTER, df['FEMALE(in thousands)'], bottom=df['MALE(in thousands)'], label='Female')
plt.legend()
plt.show()
```

# Making a histogram

```python
df=pd.read_csv("alpha_plot.csv")
plt.hist(df.y)
plt.show()
```

# Changing bins(level of detail)

```python
plt.hist(df.y, bins=40)
plt.show()
```

```python
plt.hist(df.y, range(40, 50))
plt.show()
```

# Normalization (making our histograms comparable)


```python
df=pd.read_csv('weight.csv')
plt.hist(df.male, density=True)
plt.hist(df.female, density=True)
plt.show()
```
